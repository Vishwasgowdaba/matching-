```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
def extract_dominant_color(image_path, k=1):
    # Read the image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

    # Reshape image into a 2D array of pixels
    pixels = image.reshape(-1, 3)

    # Apply KMeans clustering to find the most dominant color
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(pixels)

    # Get the dominant color
    dominant_color = kmeans.cluster_centers_[0].astype(int)
    return dominant_color  # Returns (R, G, B) color
```

```python
from IPython.display import display
from PIL import Image

# Upload an image manually in Jupyter
image_path = "images.jpg"  # Change this to your image file

# Show the uploaded image
image = Image.open("D:\images.jpeg")
display(image)

# Extract the dominant color
shirt_color = extract_dominant_color("D:\images.jpeg")
print("Shirt Dominant Color (RGB):", shirt_color)
```

```python
plt.figure(figsize=(2, 2))
plt.imshow([[shirt_color / 255]])  # Normalize to range 0-1 for Matplotlib
plt.axis("off")
plt.show()
```

```python
def color_distance(color1, color2):
    """Calculate Euclidean distance between two colors"""
    return np.linalg.norm(np.array(color1) - np.array(color2))
```

```python
pants_collection = {
    "Black Pants": [0, 0, 0],
    "Blue Jeans": [0, 0, 255],
    "Beige Trousers": [245, 245, 220],
    "Grey Pants": [128, 128, 128],
    "White Pants": [255, 255, 255]
}
```

```python
# Find the best match
best_match = min(pants_collection, key=lambda p: color_distance(shirt_color, pants_collection[p]))

print(f"Best matching pants: {best_match}")
```

```python
plt.figure(figsize=(4, 2))

# Show Shirt Color
plt.subplot(1, 2, 1)
plt.imshow([[shirt_color / 255]])  # Normalize for Matplotlib
plt.title("Shirt Color")
plt.axis("off")

# Show Best Match Pants Color
plt.subplot(1, 2, 2)
plt.imshow([[np.array(pants_collection[best_match]) / 255]])  # Normalize
plt.title(f"Best Match: {best_match}")
plt.axis("off")

plt.show()
```
